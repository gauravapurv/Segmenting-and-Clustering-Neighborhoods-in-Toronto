## Section 1 : Scraping data from the wikipedia page.

```python
# Import libraries
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
import requests  # library to handle requests
from pandas.io.json import json_normalize  # tranform JSON file into a pandas dataframe
from bs4 import BeautifulSoup
from IPython.display import display_html

!conda install -c conda-forge folium=0.5.0 --yes
import folium
from sklearn.cluster import KMeans
from IPython.display import Image 
from IPython.core.display import HTML 
```

### Scrape the data and use BeautifulSoup to create an object

```python
# Assigning the page link to variable url.
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(url, 'lxml')
tab = str(soup.table)
display_html(tab,raw=True)
```

### Creating the dataframe and cleaning the dataframe based upon "Not Assigned" neighorhood.

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

```python
# Removing rows having 'Not assigned' for borough
df = df[df.Borough != 'Not assigned']

# Combining Postal code
df = df.groupby(['Postal code','Borough'], sort=False).agg(', '.join)
df.reset_index(inplace=True)

# Replacing the name of neighbourhoods which are 'Not assigned' with names of Borough
df['Neighborhood'] = np.where(df['Neighborhood'] == 'Not assigned',df['Borough'], df['Neighborhood'])

df
```

```python
# Shape of the Dataframe.
df.shape
```

## Section - 2  - Geographical corrdinates

```python
latitude_longitude = pd.read_csv('https://cocl.us/Geospatial_data')
latitude_longitude.head()
```

### Combining the 2 dataframes with mapped potal codes

```python
latitude_longitude.rename(columns={'Postal Code':'Postal code'},inplace=True)
df1 = pd.merge(df,latitude_longitude,on='Postal code')
df1.head()
```

## Section - 3 - Clustering and Plotting of Neighorhoods in Toronto

### Retrieving data which contains Toronto in the Borough

```python
toronto = df1[df1['Borough'].str.contains('Toronto')]
toronto
```

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df1['Latitude'],df1['Longitude'],df1['Borough'],df1['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
k=5
toronto_clustering = df1.drop(['Postal code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df1.insert(0, 'Cluster Labels', kmeans.labels_)
```
